````python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
````

````python
from chat_helper import ChatHelper

helper = ChatHelper(client, model)
````

````python
messages = []

helper.add_user_message(messages, "Generate a very short event bridge rule as json")

answer = helper.chat(messages)
print(answer)
````
