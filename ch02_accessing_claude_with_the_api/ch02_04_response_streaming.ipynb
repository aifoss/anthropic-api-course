```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
from chat_helper import ChatHelper

helper = ChatHelper(client, model)
```

```python
messages = []

helper.add_user_message(messages, "Write a 1-sentence description of a fake database.")

stream = helper.chat(messages, stream=True)

for event in stream:
    print(event)
```
