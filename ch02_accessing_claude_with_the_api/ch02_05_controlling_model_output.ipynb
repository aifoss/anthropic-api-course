```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
from chat_helper import ChatHelper

helper = ChatHelper(client, model)
```

```python
# Assistant Message Prefill

messages = []

helper.add_user_message(messages, "Is tea or coffee better at breakfast?")

prefill = "Coffee is better because"
# helper.add_assistant_message(messages, prefill) ### Sonnet 4.6 model does not support assistant message prefill
helper.add_user_message(messages, prefill)

answer = helper.chat(messages)
print(answer)
```

```python
# Stop Sequences

messages = []

helper.add_user_message(messages, "Count from 1 to 10")

answer = helper.chat(messages, stop_sequences=[", 5"])
print(answer)
```
