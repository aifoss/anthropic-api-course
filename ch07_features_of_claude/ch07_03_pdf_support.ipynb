```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
from chat_helper import ChatHelper
chat_helper = ChatHelper(client, model)
```

```python
# Read PDF file, feed into Claude

import base64

with open("earth.pdf", "rb") as f:
    file_bytes = base64.standard_b64encode(f.read()).decode('utf-8')   
```

```python
messages = []

chat_helper.add_user_message(messages, [
    {
        "type": "document",
        "source": {
            "type": "base64",
            "media_type": "application/pdf",
            "data": file_bytes
        }
    },
    {
        "type": "text",
        "text": "Summarize the document in one sentence."
    }
])

response = chat_helper.chat(messages)
response
```
