```python
from dotenv import load_dotenv
load_dotenv()

import voyageai
client = voyageai.Client()
model = "voyage-3-large"
```

```python
from text_chunker import TextChunker
text_chunker = TextChunker()
```

```python
from embedding_generator import EmbeddingGenerator
embedding_generator = EmbeddingGenerator(client, model)
```

```python
from vector_index import VectorIndex
store = VectorIndex()
```

```python
with open("./report.md", "r") as f:
    text = f.read()
```

```python
# 1. Chunk the text by section

chunks = text_chunker.chunk_by_section(text)
```

```python
# 2. Generate embeddings for each chunk

embeddings = embedding_generator.generate_embedding(chunks)
```

```python
# 3. Create a vector store and add each embedding to it

for embedding, chunk in zip(embeddings, chunks):
    store.add_vector(embedding, {"content": chunk}) 
```

```python
# 4. Some time later, a user will ask a question. Generate an embedding for it

user_embedding = embedding_generator.generate_embedding("What did the software engineering dept do last year?")
```

```python
# 5. Search the store with the embedding, find the 2 most relevant chunks

results = store.search(user_embedding, 2)

for doc, distance in results:
    print(distance, "\n\n", doc["content"][0:200], "\n---\n")
```
