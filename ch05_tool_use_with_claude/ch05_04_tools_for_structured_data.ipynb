```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
from chat_helper import ChatHelper

chat_helper = ChatHelper(client, model)
```

```python
messages = []

chat_helper.add_user_message(
    messages,
    """
    Write a one-paragraph scholarly article about computer science.
    Include a title and author name.
    """    
)

response = chat_helper.chat(messages)
text = chat_helper.text_from_message(response)
print(text)
```
