```python
from dotenv import load_dotenv
load_dotenv()

import warnings
warnings.filterwarnings("ignore")

from anthropic import Anthropic
client = Anthropic()
model = "claude-opus-4-8"
```

```python
from chat_helper import ChatHelper
chat_helper = ChatHelper(client, model)
```

```python
from text_editor_tool_runner import TextEditorToolRunner
tool_runner = TextEditorToolRunner()
```

```python
# Text Edit Schema

from anthropic.types import ToolParam

text_edit_schema=ToolParam({
    "type": "text_editor_20250728",
    "name": "str_replace_based_edit_tool",
})
```

```python
# Run the conversation in a loop until the model doesn't ask for a tool use

import json

def run_conversation(messages):
    while True:
        response = chat_helper.chat(
            messages,
            tools=[text_edit_schema],
        )

        chat_helper.add_assistant_message(messages, response)
        print(chat_helper.text_from_message(response))

        if response.stop_reason != "tool_use":
            break

        tool_results = tool_runner.run_tools(response)
        chat_helper.add_user_message(messages, tool_results)

    return messages
```

```python
messages = []

chat_helper.add_user_message(
    messages,
    """
    Open the ./text_editor_tool_main.py and summarize the contents.
    """,
)

run_conversation(messages)
```

```python
messages = []

chat_helper.add_user_message(
    messages,
    """
    Open the ./text_editor_tool_main.py and write out a Python function to calculate pi to the 5th digit.
    Then create ./text_editor_tool_test.py to test your implementation.
    """,
)

run_conversation(messages)
```
